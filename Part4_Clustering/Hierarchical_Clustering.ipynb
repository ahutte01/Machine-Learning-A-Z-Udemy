# Data PreProcessing

```python
# Feb 21st 2019

# Import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import os 


# Set working directory
os.chdir('/Users/amandahutter/Documents/PythonCode/Udemy/MachineLearningA-Z/Part 4 - Clustering/Section 25 - Hierarchical Clustering') 
dataset = pd.read_csv("Mall_Customers.csv")
print(dataset.head())

# Annual Income and Spending Score are our predictive independent variables 
X = dataset.iloc[:, [3, 4]].values
```

# Use Dendogram to find Optimal Number of Clusters

```python
import scipy.cluster.hierarchy as sch
# Ward tries to minimize variance within cluster. Similar to WCSS for KMeans.
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward' ))

plt.title('Dendrogram to find optimal clusters')
plt.xlabel('All our customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

## The longest vertical line would be cut off by a horizontal line, dividing clusters into 5 groups 

# Fitting Hierarchical Clustering to Mall Dataset

```python
from sklearn.cluster import AgglomerativeClustering 
# Prepare hc algorithm 
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
Y_hc = hc.fit_predict(X)
```

# Visualise Results

```python
print(Y_hc.shape)
print(X.shape)
plt.scatter(X[Y_hc == 0, 0], X[Y_hc == 0, 1], s = 100, color = 'red', label = 'Cluster1')
plt.scatter(X[Y_hc == 1, 0], X[Y_hc == 1, 1], s = 100, color = 'blue', label = 'Cluster2')
plt.scatter(X[Y_hc == 2, 0], X[Y_hc == 2, 1], s = 100, color = 'green', label = 'Cluster3')
plt.scatter(X[Y_hc == 3, 0], X[Y_hc == 3, 1], s = 100, color = 'violet', label = 'Cluster4')
plt.scatter(X[Y_hc == 4, 0], X[Y_hc == 4, 1], s = 100, color = 'orange', label = 'Cluster5')
plt.title('Clusters of Mall Clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
