```python
# Use age and estimated salary to predict if a user will decide to buy the SUV 

# February 1st, 2019
# Logistic regression 

# Importing the libraries 
import numpy as np # included for mathematical operations
import matplotlib.pyplot as plt # plot nice charts in Python 
import pandas as pd # import datasets and manage datasets 
import os # Use this for folder and file manipulation

# Importing the dataset 
os.chdir('/Users/amandahutter/Documents/PythonCode/Udemy/MachineLearningA-Z/Part 3 - Classification/Section 14 - Logistic Regression') 
dataset = pd.read_csv('Social_Network_Ads.csv')
#print(dataset)

# We take columns 2 and 3 to get the Age & Estimated Salary for the model 
X = dataset.iloc[:, [2,3]].values # take all the rows, take all columns except the final one 
Y = dataset.iloc[:, 4].values

print(X.shape)
print(Y.shape)

# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.25, random_state = 0)

# Use Feature scaling 
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_Train = sc_X.fit_transform(X_Train)
X_Test = sc_X.transform(X_Test)
```

```python
# Fit Logistic Regression to the training data set  
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_Train, Y_Train)

# Predict the DV Test set using the classifer 
Y_Pred = classifier.predict(X_Test)
print(Y_Pred) # Predictions based on Y Test characteristics 

```

```python
# Making the confusion matrix 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true = Y_Test, y_pred = Y_Pred) # create an instance of confusion matrix class 
print(cm)
```

# 11 incorrect predictions and 89 correct predictions with the LogisticRegression model so far 

```python
# Visualising the training data set 
# Training data visualization 

# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_Set, Y_Set = X_Train, Y_Train
X1, X2 = np.meshgrid(np.arange(start = X_Set[:, 0].min() - 1, stop = X_Set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_Set[:, 1].min() - 1, stop = X_Set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
# X1 and X2 are the pixels in the meshgrid 
print(X_Set[:, 0].max()+1) # Limit 
print(X1.shape) 
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
# Apply the classifier on all the pixel points 
for i, j in enumerate(np.unique(Y_Set)):
    plt.scatter(X_Set[Y_Set == j, 0], X_Set[Y_Set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Classifier (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
